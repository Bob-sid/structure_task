```python
teams = [
    {"name": "", "wins": 10, "losses": 2, "draws": 3},
    {"name": "", "wins": 15, "losses": 1, "draws": 2},
    {"name": "", "wins": 8, "losses": 5, "draws": 3}
]

#      
best_team = max(teams, key=lambda team: team["wins"])

print(f" : {best_team['name']}  : {best_team['wins']}")
```

```python
class Team:
    def __init__(self, name, wins, losses, draws):
        self.name = name
        self.wins = wins
        self.losses = losses
        self.draws = draws

teams = [
    Team("", 10, 2, 3),
    Team("", 15, 1, 2),
    Team("", 8, 5, 3)
]

#      
best_team = max(teams, key=lambda team: team.wins)

print(f" : {best_team.name}  : {best_team.wins}")
```

```python
students = [
    {"name": " ", "course": 2, "group": "-21", "birth_year": 2003},
    {"name": " ", "course": 3, "group": "-31", "birth_year": 2002},
    {"name": " ", "course": 1, "group": "-11", "birth_year": 2004}
]

#   
youngest = max(students, key=lambda s: s["birth_year"])

print(f" : {youngest['name']} ({youngest['birth_year']} ..)")
```

```python
class Student:
    def __init__(self, name, course, group, birth_year):
        self.name = name
        self.course = course
        self.group = group
        self.birth_year = birth_year

students = [
    Student(" ", 2, "-21", 2003),
    Student(" ", 3, "-31", 2002),
    Student(" ", 1, "-11", 2004)
]

#  
youngest = max(students, key=lambda s: s.birth_year)

print(f" : {youngest.name} ({youngest.birth_year} ..)")
```

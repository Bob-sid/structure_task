```python
teams = [
    {"name": "Леви", "wins": 10, "losses": 2, "draws": 3},
    {"name": "Орли", "wins": 15, "losses": 1, "draws": 2},
    {"name": "Тигри", "wins": 8, "losses": 5, "draws": 3}
]

# Знаходимо команду з найбільшою кількістю перемог
best_team = max(teams, key=lambda team: team["wins"])

print(f"Найуспішніша команда: {best_team['name']} з перемогами: {best_team['wins']}")
```

```python
class Team:
    def __init__(self, name, wins, losses, draws):
        self.name = name
        self.wins = wins
        self.losses = losses
        self.draws = draws

teams = [
    Team("Леви", 10, 2, 3),
    Team("Орли", 15, 1, 2),
    Team("Тигри", 8, 5, 3)
]

# Знаходимо команду з найбільшою кількістю перемог
best_team = max(teams, key=lambda team: team.wins)

print(f"Найуспішніша команда: {best_team.name} з перемогами: {best_team.wins}")
```

```python
students = [
    {"name": "Іваненко Іван", "course": 2, "group": "БІ-21", "birth_year": 2003},
    {"name": "Сидоренко Олена", "course": 3, "group": "БІ-31", "birth_year": 2002},
    {"name": "Петренко Марія", "course": 1, "group": "БІ-11", "birth_year": 2004}
]

# Знаходимо наймолодшого студента
youngest = max(students, key=lambda s: s["birth_year"])

print(f"Наймолодший студент: {youngest['name']} ({youngest['birth_year']} р.н.)")
```

```python
class Student:
    def __init__(self, name, course, group, birth_year):
        self.name = name
        self.course = course
        self.group = group
        self.birth_year = birth_year

students = [
    Student("Іваненко Іван", 2, "БІ-21", 2003),
    Student("Сидоренко Олена", 3, "БІ-31", 2002),
    Student("Петренко Марія", 1, "БІ-11", 2004)
]

# Знаходимо наймолодшого
youngest = max(students, key=lambda s: s.birth_year)

print(f"Наймолодший студент: {youngest.name} ({youngest.birth_year} р.н.)")
```
